## Definitions

This is the chosen data structure that best fit the Scanflow blueprint for the following reasons:


- **Graph**: Nodes represent executors (cleaning, modeling, etc) and edges represent interactions between nodes (input/output).
- **Multi**: Multi edges are allowed. It means that there might be more than one relationship between nodes. For instance, check Tracker-Checker relationship.
- **Directed**: Dependencies are ordered, being forward or backward.
- **Lazy execution**: There are some nodes that are not executed immediately. It means, a subset of nodes waits until an action is performed. For instance, Checking and Improver modules.
- **Parent model**: Final estimator that is deployed in production.
- **Child model**: It's a copy of parent model. It aims at improving itself by a continue interaction with the Improver node.

## Formal definition

**Definition**: Scanflow is a directed labeled multigraph represented by a 7-tuple $G=(\Sigma_V, \Sigma_E, V, E, f, \ell_V, \ell_E)$. Where,

$\Sigma_V$ is the alphabet of vertices labels. $V$ is the set of nodes or vertices.

$\Sigma_V$ is the alphabet of edges labels. $E$ is the set of ordered edges.

$f: E \rightarrow V \times V=\{(u, v) | u, v \in V\}$ and $f\left(e_{1}\right)=f\left(e_{2}\right)$ in case of multiple edges.

$\ell_{V}: V \rightarrow \Sigma_{V}$, map a label for each node. For instance, preprocessing, cleaning, etc.

$\ell_{A}: A \rightarrow \Sigma_{A}$, map a label for  each edge. For instance, input, output, etc.

There are three especial nodes, $\{T, C, I\} \in V$,  Tracker (T), Checker (C) and Improver (I) which will interact to each other to enhance the parent model.

Finally, let be $\mathcal{L}$ a return function that maximizes a given metric $\theta$ over a node $n(x)$. Then, Scanflow goal is to find a 
workflow $w^*(G)$, where

$$ w^*(x) \in \operatorname*{arg\,max}_\theta \mathcal{L}(w(x)) $$

$$ w(x)=[n^{(0)}_{(x)}, n^{(1)}_{(x)},\ldots,n^{{(k-1)}}_{(x)}]= 
\begin{cases}
n^{(0)} & \mbox{ 1th node} \\
n^{(1)} & \mbox{ 2th node} \\
\vdots \\
n^{(k-1)} & \mbox{ kth node}
\end{cases}$$ 

## Properties

- With this structure we can performs a continuous improvement over a workflow, using multi edges.
- We can find neighbours of each node, as well as, sucessors and predecessor. Which are really useful to find dependecies.
- Now each edge has an identity.
- Some nodes can be lazily executed, it means, they are triggered when an action is performed. It helps to wait some actions from the Improver node, e.g., when labeling.
- Inherit graph properties allows us to use a myriad of algorithms and also perform optimization actions (maybe using a simple reinforcement learning approach).

## Examples (ongoing dev)

```python
from autodeploy.setup import Setup, Executor, Tracker, Workflow
from autodeploy.deploy import Deploy
import networkx as nx
import matplotlib.pyplot as plt

# App folder
app_dir = '/home/guess/Desktop/autodeploy/examples/demo3/data-science/'

# Workflows

executors = [Executor(name='gathering', 
                       file='gathering.py',
                       parameters={'raw_dataset': 'leaf.csv'},
                       dockerfile='Dockerfile_gathering'),
              Executor(name='preprocessing', 
                       file='preprocessing.py',
                       parameters={'training': 'training.csv',
                                   'testing': 'testing.csv'},
                       requirements='req_preprocessing.txt'),
              
              Executor(name='modeling', 
                      file='modeling.py',
                      parameters={'training': ['X_train.csv', 'y_train.csv'],
                                  'testing': ['X_test.csv', 'y_test.csv'],
                                  'model_path': 'models',
                                  'n_estimators': 3},
                      requirements='req_modeling.txt')
              ]
```

```python
G = nx.MultiDiGraph(day="Development")

# for executor in executors:
#     G.add_node(executor.name, info=executor)
    
G.add_edge(executors[0].name, executors[1].name, dependency='I/O' )
G.add_edge(executors[1].name, executors[2].name, dependency='I/O' )
# G.add_nodes_from(executors, time='Nodes')
```

```python
plt.figure(figsize =(10, 7)) 
nx.draw_networkx(G, with_label = True)
plt.title('Development')
plt.show()
```

```python
list(G.successors('preprocessing'))
```

```python
list(G.predecessors('preprocessing'))
```

### Tracker

```python
for executor in executors:
    G.add_edge(executor.name, 'Tracker', dependency='I/O' )
    
plt.figure(figsize =(10, 7)) 
nx.draw_networkx(G, with_label = True)
plt.title('Development')
plt.show()
```

### Checker

```python
G.add_edge('Tracker', 'Checker', dependency='I/O' )
G.add_edge('Checker', 'Tracker', dependency='I/O' )
    
plt.figure(figsize =(10, 7)) 
nx.draw_networkx(G, with_label = True)
plt.title('Development')
plt.show()
```

### Improver

```python
G.add_edge('Tracker', 'Improver', dependency='I/O' )
G.add_edge('Improver', 'Tracker', dependency='I/O' )
    
colors = ['blue', 'blue', 'blue', 'green', 'green', 'green']
plt.figure(figsize =(10, 7)) 
nx.draw_networkx(G, with_label = True, node_color=colors)
plt.title('Development')
plt.show()
```
